```python
import matplotlib.pyplot as plt
import numpy as np
```

## Rouge Score

### Code Llama Python 7b

```python
x = ["rouge1", "rouge2", "rougeL", "rougeLSum"]
y = [0.686, 0.573, 0.640, 0.684]

plt.bar(x, y)
plt.ylim(0.0, 1.0)
plt.xlabel("Metric")
plt.ylabel("Score")
plt.title("Fine-tuned Code Llama Python 7b rouge scores")
plt.show()
```

### Code Llama Instruct 7b

```python
x = ["rouge1", "rouge2", "rougeL", "rougeLSum"]
y = [0.511, 0.398, 0.492, 0.508]

plt.bar(x, y)
plt.ylim(0.0, 1.0)
plt.xlabel("Metric")
plt.ylabel("Score")
plt.title("Fine-tuned Code Llama Instruct 7b rouge scores")
plt.show()
```

### Comparison

```python
x = ["rouge1", "rouge2", "rougeL", "rougeLSum"]
y_base = [0.598, 0.507, 0.588, 0.598]
y_python = [0.686, 0.573, 0.640, 0.684]
y_instruct = [0.511, 0.398, 0.492, 0.508]

label_locations = np.arange(4)
width = 0.25
fig, ax = plt.subplots()

code_llama = ax.bar(label_locations - width, y_base, width, label='Base', color='orange')
code_llama_python = ax.bar(label_locations, y_python, width, label='Python', color='lightblue')
code_llama_instruct = ax.bar(label_locations + width, y_instruct, width, label='Instruct')

ax.legend([code_llama, code_llama_python, code_llama_instruct], ['Base', 'Python', 'Instruct'])

ax.set_xticks(label_locations)
ax.set_xticklabels(x)

plt.ylim(0.0, 1.0)
plt.xlabel("Metric")
plt.ylabel("Score")
plt.legend(loc=(1.02, 0.5))
plt.title("Rouge scores of the fine-tuned models")
plt.show()
```
